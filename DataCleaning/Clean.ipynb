### RFM Analysis

```python
currentDate = data['LastInteractionDate'].max()

data['recency'] = (currentDate - data['LastInteractionDate'])
data['frequecy'] = data['TotalPurchaseFrequency'] * data['Logins']
data['monetary_value'] = data['TotalPurchaseValue']


f_label = range(1, 6)
m_label = range(1, 6)
r_label = range(5, 0, -1)

data['recency'] = pd.qcut(data['recency'], 5, labels=r_label)
data['frequecy'] = pd.qcut(data['frequecy'], 5, labels=f_label)
data['monetary_value'] = pd.qcut(data['monetary_value'], 5, labels=m_label)

data[['recency', 'frequecy', 'monetary_value']] = data[['recency', 'frequecy', 'monetary_value']].astype('int64')

data['rfm_score'] = data['recency'] + data['frequecy'] + data['monetary_value']

bins = [3, 6, 9, 12, 15]
labels = ['at_risk', 'need_attention', 'loyal_customers', 'champions']
data['customer_segment'] = pd.cut(data['rfm_score'], bins=bins, labels=labels, include_lowest=True)
data.drop(columns=['recency', 'frequecy', 'monetary_value', 'rfm_score'], inplace=True)
```

### Types of encoding techniques
The choice of encoding method depends on the type of categorical data:

- For Nominal Variables (categories without a natural order):
  Method: One-Hot Encoding
  How it works: Transforms each category value into a new binary feature (0 or 1).Examples: Gender (Male/Female), Segment (A/B/C), Plan (Basic/Premium).

- For Ordinal Variables (categories with a meaningful order):
  Method: Label Encoding
  How it works: Assigns an integer to each category, preserving the intended order.Examples: Education Level (High School < Bachelor's < Master's), Satisfaction (Poor < Fair < Good).

- For High Cardinality Variables (a very large number of categories):
  Method: Target Encoding
  How it works: Replaces a category with the average value of the target variable for that category (mean encoding) or its frequency (frequency encoding). This avoids creating an excessive number of columns.



```python
# --- Define column groups ---
norminal_columns = ['Gender', 'TotalInteractionType', 'Frequency']
ordinal_columns = ['Segment', 'Plan', 'customer_segment']
cardinal_columns = ['Location', 'ProductList']

# Copy data once
encode_data = data.copy()

# --- 1. Nominal Encoding (One-Hot) ---
ohe = OneHotEncoder(sparse_output=False)
encoded_columns = []

for col in norminal_columns:
    transformed_col = ohe.fit_transform(encode_data[[col]])
    transformed_df = pd.DataFrame(
        transformed_col, 
        columns=[f"{col}_{cat}" for cat in ohe.categories_[0]],
        index=encode_data.index
    )
    encoded_columns.append(transformed_df)

encode_data = pd.concat([encode_data] + encoded_columns, axis=1)
encode_data.drop(columns=norminal_columns, inplace=True)

# --- 2. Cardinal Encoding (Mean target encoding with TotalPurchaseValue) ---
for col in cardinal_columns:
    encode_data[col] = encode_data.groupby(col)['TotalPurchaseValue'].transform('mean')

# --- 3. Ordinal Encoding (numeric mapping) ---
encoder = OrdinalEncoder()
encoded_ordinals = encoder.fit_transform(encode_data[ordinal_columns])
encode_data[ordinal_columns] = encoded_ordinals

# â Final encoded dataset
encode_data.head()
```
